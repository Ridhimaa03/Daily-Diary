```python
data={'tweets':["I love to play football.","My favourite novel is harry potter!","I hate voldemorte.","I love you!","I hate lord voldemorte.","I hate tom tiddle.","I am loving Draco Malfofy."],
      'sentiment':['positive','positive','negative','positive','negative','negative','positive']}
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.shape
```

```python
df['sentiment'].replace({'positive':1,'negative':0},inplace=True)
```

```python
df
```

```python
# Remove special characters
# Converting every thing to lower case
# Removing Stop words
# Stemming
```

```python
# converting everything to lower

def convert_lower(text):
    return text.lower()
df['tweets']=df['tweets'].apply(convert_lower)
df
```

```python
# function to remove special characters

def remove_special(text):
    x=''

    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x + ' '
    return x
```

```python
df['tweets']=df['tweets'].apply(remove_special)
```

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

nltk.download('stopwords')  # Make sure to download stopwords before using it
stopwords.words('english')
# stopwords.words('english')
```

```python
df
```

```python
def remove_stopwords(text):
    x=[]
    for i in text.split():

        if i not in stopwords.words('english'):
            x.append(i)
    y=x[:]
    x.clear()
    return y
df['tweets']=df['tweets'].apply(remove_stopwords)
df
```

```python
# Perform stemming

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
y=[]
def stem_words(text):
    for i in text:
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    return z

stem_words(['I','loved','loving','it'])
```

```python
df['tweets']=df['tweets'].apply(stem_words)
df
```

```python
def join_back(list_input):
    return " ".join(list_input)

df['tweets']=df['tweets'].apply(join_back)
df['tweets']
```

```python
# all rows of first column
X=df.iloc[:,0:1].values
X.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
X=cv.fit_transform(df['tweets']).toarray()
X.shape
```

```python
X
```

```python
X[0].mean()
```

```python
y=df.iloc[:,-1].values
y.shape
```

```python
# X,y
# Training set
# Test Set(Already know the result)
from sklearn.model_selection import train_test_split

X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf1=GaussianNB()
```

```python
clf1.fit(X_train,y_train)
```

```python
y_pred1=clf1.predict(X_test)
y_pred1
```

```python
from sklearn.metrics import accuracy_score
print("Gaussian",accuracy_score(y_test,y_pred1))
```
